```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
series = np.loadtxt("serie_menor_10mil.txt")
seg_result = pd.read_csv("file1.csv", index_col = 0)
print(seg_result)
print(series)
```

```python
means = np.zeros(0)
for i in seg_result.index:
    size = seg_result.loc[i,'finish'] - seg_result.loc[i, 'start'] + 1
    vec = np.full([size], seg_result.loc[i, 'mean'])
    means = np.concatenate((means,vec))

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
series_plot = ax1.plot(series, 'k')
ax1.plot(means, 'r-', drawstyle='steps-post', label='step-post')
```
